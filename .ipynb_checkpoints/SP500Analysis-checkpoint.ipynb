```python
import pandas as pd
import joypy
from matplotlib import pyplot as plt
import numpy as np
from collections import Counter
df = pd.read_csv("C:/Users/Mark/Documents/Stock Market Analytics/input.csv")
```

```python
#df = df.iloc[::-1]
df = df.reindex(index=df.index[::-1]).reset_index(drop = True)
df2 = df.copy(deep=True)
#df = df.sort_index(axis=1 ,ascending=True)
```

```python
#First block of code counts length of consecutive gains
Count = 0
for i in range(len(df['Change'])):  #"Change" is the name of the column in the dataframe. For loop for every change item
#for i in range(30):    
    if df['Change'][i] > 0:
        Count = Count + 1
        df.at[i,'Streak'] = Count
    else:
        Count = 0
        df.at[i,'Streak'] = Count
        if i > 0:
            if df['Streak'][(i-1)] != 0:
            #if previous streak count was 0, do nothing
            #Will keep the highest streak value and delete the previous. for example if the last value was 5, it was a steak of 5 so 4, 3, 2, 1 should be set to 0
                for j in range(2,int(df['Streak'][i - 1]+1)):
                #for j in range(1,3):
                   df.at[i-j,'Streak'] = 0 # loop as many times as the previous streak - 1
            #print(df['Streak'][i])
for i in range(len(df['Change'])): #this loop calculates the streak percent and inserts the percentage into a column
    if df['Streak'][i] == 0:
        df.at[i,'Streak Percent'] = 0
    else:
        try:
            cnt = df['Streak'][i] #Adding Try except becuase if the first piece of data is a streak, we would cause an index out of range
            df.at[i,'Streak Percent'] = round((df['Last'][i]-df['Last'][(i-cnt)])/df['Last'][(i-cnt)] *100, 3)
        except:
            df.at[i,'Streak Percent'] = N/A
            
#***********Second chunk of code is for Declines            
Count = 0
for i in range(len(df['Change'])):
#for i in range(30):    
    if df2['Change'][i] < 0:
        Count = Count + 1
        df2.at[i,'Streak'] = Count
    else:
        Count = 0
        df2.at[i,'Streak'] = Count
        if i > 0:
            if df2['Streak'][(i-1)] != 0:
            #if previous streak count was 0, do nothing
            #Will keep the highest streak value and delete the previous. for example if the last value was 5, it was a steak of 5 so 4, 3, 2, 1 should be set to 0
                for j in range(2,int(df2['Streak'][i - 1]+1)):
                #for j in range(1,3):
                   df2.at[i-j,'Streak'] = 0 # loop as many times as the previous streak - 1
            #print(df['Streak'][i])
for i in range(len(df['Change'])): #this loop calculates the streak percent and inserts the percentage into a column
    if df2['Streak'][i] == 0:
        df2.at[i,'Streak Percent'] = 0
    else:
        try:
            cnt = df2['Streak'][i]#Adding Try except becuase if the first piece of data is a streak, we would cause an index out of range
            df2.at[i,'Streak Percent'] = round((df2['Last'][i-cnt]-df2['Last'][(i)])/df2['Last'][(i-cnt)] *100, 3)
        except:
            df.at[i,'Streak Percent'] = N/A
```

```python
highest = int(max(df['Streak']))
lowest = int(min(df['Streak']))
plt.hist(df['Streak'], bins = range(1,highest+2), label = "text", edgecolor='black', align='left') #setting bins so that the histogram has "space" on the right
plt.rcParams["figure.figsize"] = (10,10)
plt.title("Consecutive Daily Advances") 
table = df['Streak'].value_counts()
plt.xlabel("Streak Length (Days)")
plt.ylabel("Number of Occurances")
missing_val = []
for k in range(highest - lowest):
    if k not in table.index:
        missing_val.append(k)  #Create array of missing value locations, for example some histogram bins will not have any values
placeholder_zeroes = [0] * len(missing_val)    
ser = pd.Series(placeholder_zeroes, index = missing_val) #Create a pandas series to append with original one
table_ex = table.append(ser)
table_ex = table_ex.reindex(index = range(highest+1))

den = sum(table_ex[1:highest+1])
name = ['']
for v in range(2,highest+2):
    val = sum(table_ex[1:v]) #Code that creates a name list for the graph percentages
    name.append("{:.1%}".format(val/den))

for idx in range(1,highest+1):
    try:
        plt.text(idx, table_ex[idx] + 5 ,name[idx], ha='center') #places name over histogram column
    except:
        plt.text(idx, 1 ,"text", ha='center')
plt.show()
```
